```python
import pandas as pd
data= pd.read_csv("/content/sample_data/Sample Dataset for Coding Task.csv")
data= data.drop(columns='Sl')
category_product_sales = data.groupby(['Category', 'Product']).sum()
category_product_sales
```

```python
category_total_sales = data.groupby('Category').sum()
category_total_sales
```

category_product_sales['Sales Percentage'] = category_product_sales['Sales Percentage'].astype(str) + '%'

```python
sales_percentage=round(category_product_sales*100/category_total_sales,2)
sales_percentage['Sales Percentage'] = sales_percentage['Sales'].astype(str) + '%'
sales_percentage.drop(columns='Sales',inplace=True)
sales_percentage
```

```python
sales_percentage.to_csv("Naga_Dheeraj_products.csv", index=True)
```
