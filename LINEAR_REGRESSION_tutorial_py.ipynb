```python
import pandas as pd
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
url  = "https://raw.githubusercontent.com/devzohaib/Simple-Linear-Regression/master/tvmarketing.csv"
data = pd.read_csv(url)

def loss_function(m,b,data):
  total_error=0
  for i in range(len(data)):
    x=data.iloc[i].TV
    y=data.iloc[i].Sales
    total_error+= (y-(m*x +b))**2
  total_error=total_error/len(data)


def gradient_descent(m_now,b_now,data,L):
  m_gradient=0
  b_gradient=0
  n=len(data)

  for i in range(n):
    x=data.iloc[i].TV
    y=data.iloc[i].Sales

    m_gradient+= -2/n * x * (y-(m_now* x +b_now))
    b_gradient+=-2/n * (y-(m_now* x +b_now))

  m=m_now - L * m_gradient
  b=b_now- L * b_gradient
  return m,b


m=0
b=0
L=0.01
number_of_iterations=100

for i in range(number_of_iterations):
  m,b=gradient_descent(m,b,data,L)

print(m,b)
m=-m
b=-b
plt.scatter(data.TV,data.Sales,color='blue')
#plt.plot(list(range(1,300)),[m*x +b for x in range(1,300)])
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

url = "https://raw.githubusercontent.com/devzohaib/Simple-Linear-Regression/master/tvmarketing.csv"
data = pd.read_csv(url)

X = data['TV'].values.reshape(-1, 1)
y = data['Sales']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = LinearRegression()
model.fit(X_train, y_train)

score = model.score(X_test, y_test)
print(round(score*100,2),"%")
```
