```python
import pandas as pd
import numpy as np
# Read the data in csv file
data = pd.read_csv('https://drive.usercontent.google.com/download?id=11GedB2pRmIpjMihPQfQhtroezpahFRB4&export=download&authuser=1')
data.head()
```

Category Mapping using the data from Table 1

```python
# Define category mappings
sex_mapping = {1: 'Male', 2: 'Female'}
age_group_mapping = {1: 'Below 22 years', 2: '22-60 years', 3: 'Above 60 years'}
education_level_mapping = {0: 'No formal education', 1: 'Primary education', 2: 'Secondary education', 3: 'Graduation and above'}

# Mapping numerical values to categorical labels
data['Sex'] = data['Sex'].map(sex_mapping)
data['Age_category'] = data['Age_category'].map(age_group_mapping)
data['Highest_education_level'] = data['Highest_education_level'].map(education_level_mapping)

data.rename(columns={'Age_group': 'Age_category'}, inplace=True) # Renaming Age_group as Age_category

data.head()# Printing the updated DataFrame
```

```python
#Finding the count of each category in the original dataset
sex_distribution = data.groupby('Sex').size() # Counts the number of Males and Females
age_group_distribution = data.groupby('Age_category').size()
education_level_distribution = data.groupby('Highest_education_level').size()
```

**random.choices()**

This function simulates random selection with weights.

*For Example:*
*It extracts category (sex) and their frequencies (weights) from a DataFrame. random.choices then picks a category based on these weights, where a higher weight increases the chance of being chosen.*

*The result, assigned to the variable sex, represents a random sex category reflecting the original data's distribution.*

```python
#  1st METHOD:

# random.choices()
'''This function simulates random selection with weights.

For Example: It extracts category (sex) and their frequencies (weights) from a DataFrame.
random.choices then picks a category based on these weights, where a higher weight increases the chance of being chosen.
The result, assigned to the variable sex, represents a random sex category reflecting the original data's distribution.'''

import random

synthetic_population = [] #Creating a empty list which is the final output

while len(synthetic_population) < 50000: # iterates untill 50,000 data are produced

    sex = random.choices(sex_distribution.index, weights=sex_distribution.values)[0]
    # Assigns a random sex based on a weighted probability distribution in 'sex_distribution'
    age_group = random.choices(age_group_distribution.index, weights=age_group_distribution.values)[0]
    # Assigns a random age_group based on a weighted probability distribution in 'age_group'
    education_level = random.choices(education_level_distribution.index, weights=education_level_distribution.values)[0]

    synthetic_population.append({'Sex': sex, 'Age_group': age_group, 'Highest_education_level': education_level})

# Creating a DataFrame named 'synthetic_population_df' from the dictionary 'synthetic_population'
synthetic_population_data = pd.DataFrame(synthetic_population)

synthetic_sex_distribution = synthetic_population_data.groupby('Sex').size() # Counts the number of Males and Females in the sex_distribution
synthetic_age_group_distribution = synthetic_population_data.groupby('Age_group').size()  # Counts the number of each description in age_group
synthetic_education_level_distribution = synthetic_population_data.groupby('Highest_education_level').size() # Counts number of each description in education_level

print(synthetic_sex_distribution)
print(synthetic_age_group_distribution)
print(synthetic_education_level_distribution)
```

```python
# Created a MultiIndex named 'major_rows' with a single level named 'Variable'
# This MultiIndex will be used as row labels

major_rows = pd.MultiIndex.from_arrays([['Sex','Age_group','Highest_education_level']],names=['Variable']) # Insert Variable column in the table

# Concatenate the dataframes along with the major rows
final_output = pd.concat([
  synthetic_population_data.groupby('Sex').size().to_frame(name='Frequency'), # Calculating the frequency distribution of 'Sex' in the synthetic population data
  # .to_frame(name='Frequency') creates a column named frequency and inserting the counts in that column
  synthetic_population_data.groupby('Age_group').size().to_frame(name='Frequency'),
  synthetic_population_data.groupby('Highest_education_level').size().to_frame(name='Frequency')], keys=major_rows) #

print(final_output)

#Save the combined distribution to a txt file
final_output.to_csv("Population_Synthesizer_.txt", index=True)
```

**BAYESIAN NETWORK**

Bayesian Network can be used to generate completely new data points that align with the relationships learned from real data.
This helps create synthetic data that reflects the complexities and dependencies found in real-world data.

```python
#  2nd METHOD:

#  BAYESIAN NETWORK METHOD
'''Bayesian Network can be used to generate completely new data points that align with the relationships learned from real data.
This helps create synthetic data that reflects the complexities and dependencies found in real-world data.'''
```

```python
import pandas as pd
import numpy as np
data = pd.read_csv('https://drive.usercontent.google.com/download?id=11GedB2pRmIpjMihPQfQhtroezpahFRB4&export=download&authuser=1')

data.rename(columns={'Age_group': 'Age_category'}, inplace=True)  # Renaming Age_group as Age_category

# Define category mappings
sex_mapping = {1: 'Male', 2: 'Female'}
age_group_mapping = {1: 'Below 22 years', 2: '22-60 years', 3: 'Above 60 years'}
education_level_mapping = {0: 'No formal education', 1: 'Primary education', 2: 'Secondary education', 3: 'Graduation and above'}

# Mapping numerical values to categorical labels
data['Sex'] = data['Sex'].map(sex_mapping)
data['Age_category'] = data['Age_category'].map(age_group_mapping)
data['Highest_education_level'] = data['Highest_education_level'].map(education_level_mapping)
```

```python
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import BayesianEstimator
from pgmpy.sampling import BayesianModelSampling

# Created a Bayesian Network model with directed dependencies
model = BayesianNetwork([('Sex', 'Age_category'), ('Age_category', 'Highest_education_level')])# Sex influences Age_category & Age_category influences Highest_education_level

# Estimating the parameters of the Bayesian network from the dataset
model.fit(data, estimator=BayesianEstimator)

# Sampling from the Bayesian network to generate the population
sampling = BayesianModelSampling(model)

# Generating 50,000 samples from the Bayesian network model
synthesized_population = sampling.forward_sample(size=50000)
```

```python
sex_counts = synthesized_population['Sex'].value_counts() # Counts the number of Males and Females in the synthetic_population
age_group_counts = synthesized_population['Age_category'].value_counts()
education_level_counts = synthesized_population['Highest_education_level'].value_counts()
```

```python
print(sex_counts)
print(age_group_counts)
print(education_level_counts)
```

```python
# Creating a MultiIndex named 'major_rows' with a single level with column name as 'Variable'
major_rows = pd.MultiIndex.from_arrays([['Sex','Age_group','Highest_education_level']],names=['Variable'])

final_output = pd.concat([sex_counts ,age_group_counts,education_level_counts], keys=major_rows) # Concatenating frequency of each Variable into a tabular column
print(final_output)
# Save the combined distribution to a txt file
final_output.to_csv("Bayesian_Population_Synthesizer.txt", index=True)
```
