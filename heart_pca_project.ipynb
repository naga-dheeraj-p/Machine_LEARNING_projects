```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/sample_data/heart.csv')
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#removing outliners based on z score
z_score=np.abs((data-data.mean())/data.std())
z_score.head()
```

```python
new_data=z_score.loc[:,(z_score<3).all()]
new_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
data=pd.DataFrame(data)
label_encoder=LabelEncoder()
data['sex_LabeEncoded']=label_encoder.fit_transform(data['Sex'])
data['ChestPainType_LabelEncoded']=label_encoder.fit_transform(data['ChestPainType'])
data['RestingECG_LabelEncoded']=label_encoder.fit_transform(data['RestingECG'])
data['ST_Slope_LabeEncoded']=label_encoder.fit_transform(data['ST_Slope'])
data['ExerciseAngina_LabeEncoded']=label_encoder.fit_transform(data['ExerciseAngina'])
a=data.drop(['Sex','ChestPainType','RestingECG','ST_Slope','ExerciseAngina'],axis=1)
a
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
a_scaled=scaler.fit_transform(a)
a_scaled
```

```python
from sklearn.model_selection import train_test_split
y=data['HeartDisease']
x_train,x_test,y_train,y_test=train_test_split(a_scaled,y,test_size=0.2,random_state=100)
```

```python
#SVM
from sklearn import svm
classifier=svm.SVC(kernel='linear',gamma='auto',C=2)
classifier.fit(x_train,y_train)
y_predict=classifier.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_predict)
print(accuracy*100,'%')
```

```python
#LogisticRegression
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
#model.score(x_test,y_test)
y_predict=model.predict(x_test)
accuracy=accuracy_score(y_test,y_predict)
print(accuracy*100,'%')
```

```python
#RandomForest
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier()
rf_model.fit(x_train,y_train)
y_rf_predict=rf_model.predict(x_test)
accuracy=accuracy_score(y_test,y_rf_predict)
print(accuracy*100,'%')
```

```python
# PCA
from sklearn.decomposition import PCA
pca=PCA(0.9)
x_pca=pca.fit_transform(a_scaled)
pca.explained_variance_ratio_
```

```python
print(pca.n_components_)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
pca=PCA(0.9)
x_pca=pca.fit_transform(a_scaled)
pca.explained_variance_ratio_
y=data['HeartDisease']
x_train_pca,x_test_pca,y_train,y_test=train_test_split(x_pca,y,test_size=0.2,random_state=100)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train_pca,y_train)
#model.score(x_test,y_test)
y_predict=model.predict(x_test_pca)
accuracy=accuracy_score(y_test,y_predict)
print(accuracy*100,'%')
```
